# Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

# Original Image

```python
image = cv2.imread("PK.jpeg")
plt.imshow(image[:,:,::-1])
```

# Converting to Grayscale

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
```

# Inverting the Grayscale Image

```python
inverted_image = cv2.bitwise_not(gray_image)
plt.imshow(inverted_image)
```

# Blurring the image using Gaussian function

```python
blurred = cv2.GaussianBlur(inverted_image, (111,111),0)
plt.imshow(blurred)
```

```python
inverted_blurred=cv2.bitwise_not(blurred)
plt.imshow(inverted_blurred)
```

# Pencil Sketch

```python
sketch_image=cv2.divide(gray_image,inverted_blurred, scale=256.0)
pencil_sketch=cv2.cvtColor(sketch_image, cv2.COLOR_BGR2RGB)
plt.imshow(pencil_sketch)
plt.axis('off')
plt.show()
```
